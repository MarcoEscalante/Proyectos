```python
def modelo_prediccion(x, y, epochs, imprimir_error_cada, escalar_lr):
    uno = np.ones_like(x)
    uno_t = np.vstack((x,uno))
    inicio = np.array([0.,0.])
    error = np.ones(epochs)
    modelo_resul={}
    for i in range(epochs):
        yhat = np.matmul(inicio,uno_t)
        error[i] = 0.5*(np.mean((y - yhat)**2))
        gradiante_m = np.mean((yhat - y)*x)
        gradiante_b = np.mean(yhat - y)
        gradiante = np.array([gradiante_m, gradiante_b])
        inicio -= escalar_lr*gradiante
        modelo_resul[i] = inicio.copy()
        if i % imprimir_error_cada == 0:
            print("%i Iteracion --> Funcion de costo = %0.1f" %(i, error[i]))
    return modelo_resul, error
```
